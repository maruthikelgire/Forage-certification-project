```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Delinquency_prediction_dataset_csv.csv')

# Display the first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Display basic information about the dataset
print("\nDataset Information:")
df.info()

# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("Missing values before cleaning:")
print(df.isnull().sum())

# Impute missing values with the median
for col in ['Income', 'Credit_Score', 'Loan_Balance']:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Standardize 'Employment_Status'
df['Employment_Status'] = df['Employment_Status'].replace({'EMP': 'Employed', 'employed': 'Employed'})

# Check for and remove duplicates
print(f"\nNumber of duplicate rows before removal: {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"Number of duplicate rows after removal: {df.duplicated().sum()}")


# Verify cleaning
print("\nMissing values after cleaning:")
print(df.isnull().sum())

print("\nUnique values in 'Employment_Status' after cleaning:")
print(df['Employment_Status'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the visualizations
sns.set_style('whitegrid')

# Create histograms for numerical features
numerical_features = ['Age', 'Income', 'Credit_Score', 'Credit_Utilization', 'Loan_Balance', 'Debt_to_Income_Ratio', 'Account_Tenure']
df[numerical_features].hist(bins=20, figsize=(15, 10))
plt.suptitle('Distribution of Numerical Features', y=1.02, size=16)
plt.tight_layout()
plt.savefig('numerical_features_distribution.png')
plt.show()
```
